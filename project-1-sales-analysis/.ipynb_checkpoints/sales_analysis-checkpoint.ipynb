```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_excel("data/superstore.csv")
df.head()
```

```python
# Check the first 5 rows
df.head()
```

```python
# Get info about columns, types, and missing values
df.info()
```

```python
# Summary statistics of numeric columns
df.describe()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.shape
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
```

```python
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
```

```python
df.info()
df.head()
```

```python
total_sales = df['Sales'].sum()

print("Total Sales:", total_sales)
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print(region_sales)

region_sales.plot(kind='bar', figsize=(7,5), color='skyblue')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print(category_sales)

category_sales.plot(kind='bar', figsize=(7,5), color='orange')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_products)

top_products.plot(kind='barh', figsize=(8,5), color='green')
plt.title("Top 10 Products by Sales")
plt.xlabel("Sales")
plt.ylabel("Product Name")
plt.show()
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.month)['Sales'].sum()
print(monthly_sales)

monthly_sales.plot(kind='line', marker='o', figsize=(8,5))
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

# Insights from Sales Analysis

- **Region with highest sales:** West
- **Most selling category:** Technology

- **Monthly sales trend:** Peak sales in November, lowest in February
